sklearn.datasets.load_iris: Loads the Iris dataset.
train_test_split: Splits data into training and testing sets.
DecisionTreeClassifier: A simple tree-based classifier.
accuracy_score, confusion_matrix: Metrics to evaluate model performance.
matplotlib.pyplot, seaborn: Libraries for plotting and visualization.

```python

# Step 1: Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Step 2: Load build-in dataset
iris = load_iris()
X = iris.data
y = iris.target
```

Splits the dataset into:

70% training data
30% testing data

random_state=42: Ensures reproducibility.

clf: The classifier object.
.fit(): Trains the model using training data.

```python
# Step 3: Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

clf: The classifier object.
.fit(): Trains the model using training data.

```python
# Step 4: Train a Decision Tree Classifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

Uses the trained model to predict labels for the test set.

accuracy_score: Measures how many predictions were correct.
confusion_matrix: Shows how predictions compare to actual labels.
Prints the accuracy and confusion matrix.

```python
# Step 5: Make predictions
y_pred = clf.predict(X_test)
```

Creates a heatmap of the confusion matrix.
annot=True: Shows numbers in the boxes.
xticklabels and yticklabels: Label the axes with species names.

```python

# Step 6: Evaluate the model
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {acc:.2f}")
print("Confusion Matrix:")
print(cm)
```

```python

# Step 7: Visualize the confusion matrix
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
